```python
dang_df = pd.read_csv("dang_menu.csv")
```

```python
dang_df
```

```python
dang_df = dang_df.drop(columns=['Unnamed: 0', 'index'])
```

```python
menu_df = buu_df.append(dang_df, ignore_index=True)
menu_df
```

```python
menu_df.drop_duplicates(keep='first',ignore_index=True)
```

```python
def discount(x):
    if type(x)==float and pd.isna(x):
        return x
    else:
        k = json.loads(x.replace("'", '"'))
        return k["value"]
menu_df['discount_price']=menu_df['discount_price'].apply(lambda x: discount(x))
```

```python
menu_df = menu_df.drop(columns=["unit", "label", "discount_remaining_quantity", 
                               "is_group_discount_item", "display_order", "period",
                               "is_searchable", "limit_per_order", "limit_type", "available_time"])
```

```python
menu_df.to_csv("menu_preprocessed.csv", index=False)
```
