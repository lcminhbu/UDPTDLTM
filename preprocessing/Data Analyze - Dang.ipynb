### 1. Món ăn có điểm trung bình cao nhất theo quận

```python
infor_dataframe
```

```python
unique_list_food = []
for index_1 in infor_dataframe["serves"].tolist():
    check = str(index_1).split(',')
    for checked in check:
        if checked.strip() not in unique_list_food:
            unique_list_food.append(checked.strip())
```

```python
unique_district = infor_dataframe["district"].unique().tolist()
value_0 = []
for index in range(len(unique_district)):
    value_0.append(list(range(len(unique_list_food))))
```

```python
ques_1 = pd.DataFrame(value_0,columns=unique_list_food)
for col in ques_1.columns:
    ques_1[col].values[:] = 0.0
```

```python
ques_1.index = unique_district
```

```python
ques_1 = ques_1.astype(float)
ques_1.dtypes
```

```python
infor_dataframe.loc[infor_dataframe['average_score']=='_._', 'average_score'] = np.nan
```

```python
for district in unique_district:
    for name in unique_list_food:
        point = infor_dataframe[(infor_dataframe["district"] == district) & (infor_dataframe["serves"]\
                                                        .str.contains(name))]["average_score"]
        point = np.array(point.tolist(), dtype=np.float32).mean()
        ques_1.loc[district][name] = float(point)
        
```

```python
for index in ques_1.index.tolist():
    print(f"{index} - {ques_1.loc[index].idxmax()}")
```
