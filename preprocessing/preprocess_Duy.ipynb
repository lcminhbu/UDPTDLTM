## Thống kê dữ liệu

Về dữ liệu Restaurant

```python
restaurant_info_df.groupby('district').size()
```

Ta thấy dữ liệu của chúng ta xuất hiện 2 quận: `Quận 2`, `Quận 9`. 2 quận này hiện giờ đã được sát nhập vào thành phố Thủ Đức nên để đồng nhất ta sẽ chuyển cột `district` ứng với 2 quận này thành `Tp. Thử Đức` để thống nhất.

```python
restaurant_info_df['district'] = restaurant_info_df['district'].replace('Quận 2', 'Tp. Thủ Đức')
restaurant_info_df['district'] = restaurant_info_df['district'].replace('Quận 9', 'Tp. Thủ Đức')
restaurant_info_df.groupby('district').size()
```

```python
import matplotlib.pyplot as plt
# Figure Size
fig, ax = plt.subplots(figsize =(16, 9))
 
# Horizontal Bar Plot
ax.barh(restaurant_info_df['district'].unique(), restaurant_info_df.groupby('district').size())
 
# Remove axes splines
for s in ['top', 'bottom', 'left', 'right']:
    ax.spines[s].set_visible(False)
 
# Remove x, y Ticks
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')
 
# Add padding between axes and labels
ax.xaxis.set_tick_params(pad = 5)
ax.yaxis.set_tick_params(pad = 10)
 
# Add x, y gridlines
ax.grid(b = True, color ='grey',
        linestyle ='-.', linewidth = 0.5,
        alpha = 0.2)
 
# Show top values
ax.invert_yaxis()
 
# Add annotation to bars
for i in ax.patches:
    plt.text(i.get_width()+0.2, i.get_y()+0.5,
             str(round((i.get_width()), 2)),
             fontsize = 10, fontweight ='bold',
             color ='grey')
 
# Add Plot Title
ax.set_title('Amount of restaurants in each district',
             loc ='left', )
 
# Add Text watermark
fig.text(0.9, 0.15, 'Jeeteshgavande30', fontsize = 12,
         color ='grey', ha ='right', va ='bottom',
         alpha = 0.7)
 
# Show Plot
plt.show()
```

Dựa vào biểu đồ ta có thể thấy số lượng hàng quán ở quận 10 là nhiều nhất, tiếp đến là quận 3, quận 7, quận 5.  
Mà hiện tại do số lượng cửa hàng đang hơi nhỏ nên thống kê này có vẻ chưa chuẩn lắm. Cần thu thập thêm nhiều quán nữa.
