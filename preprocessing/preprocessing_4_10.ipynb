```python
import pandas as pd
```

# Khám phá dữ liệu 1 

File dữ liệu store_info_fixed_29-9.txt

```python
store_info_df = pd.read_csv('store_info_fixed_29-9.txt',delimiter=',',index_col='idNum')
store_info_df.head()
```

## Thông tin chung về dữ liệu 

```python
# Số dòng và cột
print('Số dòng và số cột: {0} {1}'.format(store_info_df.shape[0],store_info_df.shape[1]))
# Có dòng bị lặp không
print('Số dòng bị lặp: ', store_info_df.duplicated().sum())
```

Mỗi dòng trong dữ liệu có các thông tin về quán ăn như: Tên quán, địa chỉ, số điểm đánh giá theo tiêu chí, ...

Kiểu dữ liệu cho từng cột. Nhìn qua thì các điểm đánh giá là kiểu float64. Tuy nhiên một vài cột là kiểu dữ liệu dạng số nhưng hiện tại đang có kiểu dữ liệu là object như: review_count, average_score, views. Chúng ta cần lưu ý những cột này trong những bước sau. 

```python
store_info_df.dtypes
```

Kiểm tra thông tin của các cột review_count, average_score, views

```python
store_info_df[['review_count', 'average_score', 'views']].describe()
```

```python
store_info_df['review_count'].unique()
```

```python
store_info_df['review_count'] = store_info_df['review_count'].str.replace(',','').astype('int32')
store_info_df['review_count'].unique()
```

```python
store_info_df['average_score'] = store_info_df['average_score'].str.replace('_._','nan',regex=False).astype('float32')
store_info_df['average_score'].unique()
```

```python
a = store_info_df['views'].str.extract(r'(\d+.\d+)K')
b = store_info_df['views'].str.extract(r'(\d+)')
a = a + '00'
a = a.fillna(b)
store_info_df['views'] = a[0].str.replace('.','',regex=False)
store_info_df['views'] = store_info_df['views'].astype('int64')
```

Kiểm tra lại thông tin của 3 cột review_count, average_score, views

```python
 store_info_df[['review_count', 'average_score', 'views']].describe()
```

- Các cột dữ liệu dạng số là: 

` 'review_count', 'average_score', 'space_score', 'position_score', 'quality_score','serve_score','price_score','views'`

- Các cột dữ liệu dạng phân loại là:

` '_id', 'name', 'address', 'district', 'opening_time', 'suitable_time', 'last_pick_up_time',  'prepare_time', 'holiday', 'type', 'style', 'serves', 'suitable_with', 'available', 'capacity', 'serves',  'price_range' `

- Các cột dữ liệu còn lại:
` 'food_list', 'comment_list', 'food_link', 'parking_lots', 'website'  ` (Chúng ta sẽ xử lý những dữ liệu này khi cần đến)

Kiểm tra lại thông tin của những cột dữ liệu trên

```python
store_info_df[\
['review_count', 'average_score', 'space_score', 'position_score', 'quality_score','serve_score','price_score','views']]\
.describe()
```

Nhìn qua thì cột điểm đánh giá như ` 'space_score', 'position_score', 'quality_score', 'serve_score', 'price_score' ` có điểm đánh giá nhỏ nhất là -1. Chúng ta sẽ nhìn qua những dòng này.

```python
pd.set_option('display.max_colwidth', None)
store_info_df[(store_info_df['space_score'] == -1) |\
             (store_info_df['position_score'] == -1) |\
             (store_info_df['quality_score'] == -1) |\
             (store_info_df['serve_score'] == -1) |\
             (store_info_df['price_score'] == -1)][['name','food_link','space_score',\
                                                    'position_score', 'quality_score', 'serve_score', 'price_score']]
```

Có 160 dòng có những giá trị -1 như thế này.

Qua kiểm tra địa chỉ của những quán này trên shopee thì những giá trị -1 này đại diện cho giá trị rỗng.

```python
store_info_df[['name', 'address', 'district', 'opening_time', 'suitable_time',\
    'last_pick_up_time',  'prepare_time', 'holiday', 'type', 'style', 'serves',\
    'suitable_with', 'available','capacity', 'serves',  'price_range']].describe()
```

```python
store_info_df[['name', 'address', 'district', 'opening_time', 'suitable_time',\
    'last_pick_up_time',  'prepare_time', 'holiday', 'type', 'style', 'serves',\
    'suitable_with', 'available','capacity', 'serves',  'price_range']].info()
```
